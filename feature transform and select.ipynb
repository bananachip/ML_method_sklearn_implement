# 特征变换 和 特征选择


```python
from sklearn.datasets import load_iris
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
iris
iris.feature_names
print(iris.data)
print(iris.target)
iris.target_names
```

## 数据预处理 preprocessing库

### 1.无量纲化：标准化（StandardScaler），区间缩放法(MinMaxScaler)

```python
from sklearn.preprocessing import StandardScaler
standard = StandardScaler()
iris_standard = standard.fit_transform(iris.data)
iris_standard

plt.plot(iris_standard[:,0])
```

```python
from sklearn.preprocessing import MinMaxScaler
iris_minmax = MinMaxScaler().fit_transform(iris.data)
iris_minmax

plt.plot(iris_minmax[:,0])
```

```python
from sklearn.preprocessing import Normalizer
iris_norm = Normalizer(norm='l2').fit_transform(iris.data)
iris_norm

plt.plot(iris_norm[:,0])
```

```python
#二值化
from sklearn.preprocessing import Binarizer
Binarizer(threshold = 3).fit_transform(iris.data)
```

```python
#哑编码
#print(iris.target.reshape((-1,1)))  #-1表示自动推算得到
from sklearn.preprocessing import OneHotEncoder
OneHotEncoder().fit_transform(iris.target.reshape((-1,1))).toarray()
```

```python
#缺失值计算
from numpy import vstack,array,nan
print (vstack((array([nan,nan,nan,nan]),iris.data)))

from sklearn.preprocessing import Imputer
Imputer(strategy='most_frequent',missing_values=nan).fit_transform(vstack((array([nan,nan,nan,nan]),iris.data)))
```

```python
#数据变换 多项式特征变换
from sklearn.preprocessing import PolynomialFeatures
PolynomialFeatures(degree = 2).fit_transform(iris.data)
```

```python
#数据变换 单变元函数变换
from numpy import log1p
from sklearn.preprocessing import FunctionTransformer
FunctionTransformer(log1p).fit_transform(iris.data)
```

# 特征选择


## Filter过滤法

```python
#方差选择法
from sklearn.feature_selection import VarianceThreshold
VarianceThreshold(threshold=3).fit_transform(iris.data)
#返回值为特征选择后的数据
```

```python
#相关系数法
#第一个参数为计算评估特征是否好的函数，该函数输入特征矩阵和目标向量，
#输出二元组（评分，P值）的数组，数组第i项为第i个特征的评分和P值。在此定义为计算相关系数
from sklearn.feature_selection import SelectKBest
from scipy.stats import pearsonr #皮尔逊相关系数

SelectKBest(lambda X, Y: array(map(lambda x:pearsonr(x, Y), X.T)).T, k=2).fit_transform(iris.data, iris.target)
```

```python
#pipeline
from sklearn import svm
from sklearn.datasets import samples_generator
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from sklearn.pipeline import Pipeline
# generate some data to play with
X, y = samples_generator.make_classification(n_informative=5, n_redundant=0, random_state=42)
 # ANOVA SVM-C
anova_filter = SelectKBest(f_regression, k=5)
clf = svm.SVC(kernel='linear')
anova_svm = Pipeline([('anova', anova_filter), ('svc', clf)])
# You can set the parameters using the names issued
# For instance, fit using a k of 10 in the SelectKBest
# and a parameter 'C' of the svm
anova_svm.set_params(anova__k=10, svc__C=.1,svc__kernel='linear').fit(X, y)
prediction = anova_svm.predict(X)
anova_svm.score(X, y) 
 # getting the selected features chosen by anova_filter
anova_svm.named_steps['anova'].get_support()
```
